```python
import cv2
import os
import glob
import numpy as np

# Plotting
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")
plt.rcParams["font.family"] = "Arial"

import torch
```

# Visualize training data results

```python
# With class weights in criterion
path = '/home/dieumynguyen/BeeSegmentation/data/processed/training_metrics.pt'
```

```python
xxx = torch.load(path)
```

```python
len(xxx['losses']['train'])
```

```python
train_accs = xxx['accs']['train']
train_mious = xxx['mious']['train']
train_losses = xxx['losses']['train']
train_tprs = xxx['true_positive_rates']['train']
```

```python
val_accs = xxx['accs']['val']
val_mious = xxx['mious']['val']
val_losses = xxx['losses']['val']
val_tprs = xxx['true_positive_rates']['val']
```

```python
plt.plot(train_losses, label='Train')
plt.plot(val_losses, label='Validation')
plt.legend()
plt.title('Loss');
```

```python
plt.plot(train_accs, label='Train')
plt.plot(val_accs, label='Validation')
plt.legend()
plt.title('Accuracy');
```

```python
plt.plot(train_mious, label='Train')
plt.plot(val_mious, label='Validation')
plt.legend()
plt.title('mIOUs');
```

```python
max(xxx['true_positive_rates']['train']), max(xxx['true_positive_rates']['val'])
```

```python
plt.plot(train_tprs, label='Train')
plt.plot(val_tprs, label='Validation')
plt.legend()
plt.title('TPRs');
plt.ylim(0, 1.0);
```
