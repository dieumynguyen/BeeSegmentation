```python
def _finalize_mask(mask, n_classes):
    new_mask = []
    for class_i in range(n_classes):
        new_mask_i = np.zeros_like(mask)
        new_mask_i[mask==class_i] = 1
        new_mask.append(new_mask_i)
    new_mask = np.array(new_mask)
    return new_mask
```

```python
def _finalize_mask(mask, n_classes):
    new_mask = np.zeros((n_classes, *mask.shape))
    for i in range(n_classes):
        new_mask[i][np.where(mask==i)] = 1
    return new_mask
```

```python
new_mask = _finalize_mask(mask, 3)
```
